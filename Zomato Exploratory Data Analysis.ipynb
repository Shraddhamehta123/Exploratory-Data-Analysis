## Importing the required datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the data

```python
df=pd.read_csv('zomato.csv', error_bad_lines=False, engine='python')
```

```python
df.head()
```

## Count no. of rows and columns

```python
df.shape
```

## To see the columns list

```python
df.columns
```

## Dropping the not required columns

```python
df=df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1)
df.head()
```

## To fetch information about the data

```python
df.info()
```

## Dropping duplicates

```python
df.drop_duplicates(inplace=True)
df.shape
```

## Cleaning Rate column

```python
print(df['rate'].unique())
```

### From the above result-
### 'NEW' and '-' should be replaced with 'nan'
### '/5' should be removed from all values

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate']=df['rate'].apply(handlerate)
df['rate'].head()
```

### Filling null values in Rate column with 'mean value' as there are lot of null

```python
df['rate'].fillna(df['rate'].mean(), inplace=True)
df['rate'].isnull().sum()
```

### Cleaning Location Column and listed_in(city) Column

### Drop all rows with null values

```python
df.dropna(inplace=True, how='all')
df.shape
```

```python
df.rename(columns={'approx_cost(for two people)':'Cost2plates','listed_in(type)':'Type'}, inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

### As both listed_in(city) and location columns are simillar, lets keep only location column

```python
df=df.drop(['listed_in(city)'], axis=1)
```

### Putting location value as 'others' for locations where restaurant count is less than 300

```python
df['location'].value_counts()
```

```python
location=df['location'].value_counts(ascending=False)
location_lessthan300=location[location<300]
def handle_location(value):
    if (value in location_lessthan300):
        return 'others'
    else:
        return value
df['location']=df['location'].apply(handle_location)
df['location'].value_counts()
```

## Cleaning Cost2plates column

```python
df['Cost2plates'].unique
```

### Removing ',' from Cost2plates column

```python
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
df['Cost2plates']=df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique
```

## Cleaning rest_types column 
### Changing 'rest_types' values having count of restaurants less than 1000 to 'others'

```python
df.head()
rest_types=df['rest_type'].value_counts(ascending=False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handle_rest_type(value):
    if (value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
df['rest_type']=df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

## Cleaning cuisine column

```python
cuisines=df['cuisines'].value_counts(ascending=False)
cuisines_lessthan100=cuisines[cuisines<100]
def handle_cuisines(value):
    if (value in cuisines_lessthan100):
        return 'others'
    else:
        return value
df['cuisines']=df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
